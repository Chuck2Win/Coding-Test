# Back Tracking : 제약조건에 부합하는 해 찾기
## 상태 공간 트리, DPS, Promising/Prunning


```python
# N queen 문제
```

```python
def solution(N):
    
    final = []
    dfs(N,0,[],final)
    return final
```

```python
def dfs(N, current_row, candi, final):
    if current_row == N:
        final.append(candi)
        return
    else:
        for col in range(N): # 열
            if is_available(col,candi):
                dfs(N,current_row+1,candi+[col],final)                
```

```python
# 안됨

def is_available(col,candi):
    '''
    example : i = 2, candi =[1,3] -> current : 3 -> index론 2
    대각선 체크, 수직 체크
    '''
    current = len(candi) 
    for j in range(current): # j : 행
        #print((candi[j] == i) and (abs(candi[j]-i)==(current-j)))
        if (candi[j] == i) and (abs(candi[j]-i)==(current-j)): # 열비교, 
            return False
    return True
```

```python
# debugging

def is_available(col,candi):
    '''
    example : i = 2, candi =[1,3] -> current : 3 -> index론 2
    대각선 체크, 수직 체크
    '''
    current_row = len(candi) 
    for row in range(current_row): # j : 행
        if (candi[row] == col) or (abs(candi[row]-col)==(current_row-row)): # 열비교, -> or..
            return False
    return True
```
