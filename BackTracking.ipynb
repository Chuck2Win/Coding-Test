```python
def solution(N):
    
    final = []
    dfs(N,0,[],final)
    return final
```

```python
def dfs(N, current_row, candi, final):
    if current_row == N:
        final.append(candi)
        return
    else:
        for col in range(N): # 
            if is_available(col,candi):
                dfs(N,current_row+1,candi+[col],final)                
```

```python
# 

def is_available(col,candi):
    '''
    example : i = 2, candi =[1,3] -> current : 3 -> index 2
     ,  
    '''
    current = len(candi) 
    for j in range(current): # j : 
        #print((candi[j] == i) and (abs(candi[j]-i)==(current-j)))
        if (candi[j] == i) and (abs(candi[j]-i)==(current-j)): # , 
            return False
    return True
```
